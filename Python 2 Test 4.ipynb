```python
#1:Take two numbers, start and end, and print all prime numbers in that range.

start=eval(input("Enter starting number:"))
end=eval(input("Enter ending number:"))

print("Prime numbers between", start ,"and" ,end, "are:")

for num in range(start, end + 1):
   if num > 1:
       for i in range(2, int(num**0.5) + 1):
           if (num % i) == 0:
               break
       else:
           print(num, end=" ")
```

```python
#3Take a number and calculate the sum of the factorials of its digits.

import math

num = input("Enter a number: ")
total_sum = sum(math.factorial(int(digit)) for digit in num)
print(f"The sum of the factorials of the digits of {num} is: {total_sum}")


```

```python
#4The computer chooses a random number between 1 and 50. User keeps guessing until correct. Give hints if the guess is higher or lower.

import random
r=random.randint(1, 50)
n=0
while  n != r:
    n=eval(input("Guess the number:"))
    if n < r:
        print("Guess Higher,Try Again")
    elif n > r:
        print("Guess Lower,Try Again")
    else:
         print("Correct Guessed")
else:
    print("Game Over")
```

```python
#5 Take an integer and count how many digits are even and how many are odd

num_str = input("Enter an integer: ")

even_count = 0
odd_count = 0

for digit in num_str:
    if digit.isdigit():  
        n = int(digit)
        if n % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

print(f"Even digits: {even_count}")
print(f"Odd digits: {odd_count}")
```

```python
#6 Take a string and a character, and count occurrences of that character without using built-in functions.

text = input("Enter a string: ")
char_to_find = input("Enter the character to count: ")


count = 0

for char in text:
    if char == char_to_find:
        count += 1

print(f"The character '{char_to_find}' appears {count} times.")
```

```python
#7 Take N numbers as input and print the multiplication table for each number

n = int(input("How many numbers do you want to enter? "))

numbers = []
for i in range(n):
    val = int(input(f"Enter number {i+1}: "))
    numbers.append(val)

print("-" * 20)
for num in numbers:
    print(f"Multiplication Table for {num}:")
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")
```

```python
# 8 Take a number and sum all digits that are prime (2, 3, 5, 7).
num = input("Enter a number: ")

prime_digits = "2357"

prime_sum = 0

for digit in num:
    if digit in prime_digits:
        prime_sum += int(digit)

print(f"The sum of prime digits is: {prime_sum}")
```

```python
#9 Print a diamond-shaped star pattern for a given odd number N.
n = int(input("Enter an odd number: "))

if n % 2 == 0:
    print("Please enter an odd number for a perfect diamond.")
else:
    for i in range(1, n + 1, 2):
        spaces = (n - i) // 2
        print(" " * spaces + "*" * i)

    for i in range(n - 2, 0, -2):
        spaces = (n - i) // 2
        print(" " * spaces + "*" * i)
```

```python
#10 Print a rectangle of size M × N with stars, but replace the diagonal with numbers 1 to min(M,N).
m = int(input("Enter number of rows (M): "))
n = int(input("Enter number of columns (N): "))

for i in range(m):
    for j in range(n):
        if i == j:
            print(i + 1, end=" ")
        else:
            print("*", end=" ")
    print()
```
