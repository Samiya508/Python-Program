```python
#1 Name Shortener AppInput a full name and output initials (e.g., Ravi Kishan Patel → R.K.P.).

full_name = input("Enter the full name :- ")

ans = " ".join([i[0]+"." for i in full_name.split()])
print(ans)
```

```python
#2 Restaurant Menu System Use a dictionary for items and prices. Ask user to order items, then display total cost.

menu = {
    "Burger": 150,
    "Pizza": 350,
    "Fries": 90,
    "Soda": 50,
    "Ice Cream": 120
}

print("--- MENU ---")
for item, price in menu.items():
    print(f"{item}: ₹{price}")
print("------------")

total_cost = 0
ordering = True

while ordering:
    user_choice = input("\nEnter the item you want to order (or type 'done' to finish): ").strip().title()
    
    if user_choice == 'Done':
        ordering = False
    elif user_choice in menu:
        total_cost += menu[user_choice]
        print(f"Added {user_choice} to your order. Current total: ₹{total_cost}")
    else:
        print("Sorry, that item is not on the menu. Please try again.")

print(f"\nYour final bill amount is: ₹{total_cost}")
print("Thank you for your order!")
```

```python
#3 Email Validator Check if an email contains '@' and ends with '.com'. Print “Valid” or “Invalid”.
e = input("Enter the email :-")

if "." in e:
    a = e.index(".")

print("Email Id :-",e)
if "@" in e and e[a:] == ".com":
    print("Valid Email")

else:
    print("Invalid Email")
```

```python
#4 Unique Words Extractor Input a sentence, convert to lowercase, and print unique words using a set.
sentence = input("Enter a sentence: ")

w = sentence.lower().split()

cleaned_words = [word.strip(".,!?;:()\"'") for word in w]

unique_words = set(cleaned_words)

print("\nUnique words found:")
for word in unique_words:
    if word:  # Filter out empty strings from standalone punctuation
        print(f"- {word}")
```

```python
#5 Class Marks Tracker Input names and marks of 5 students, store in a dictionary, and print the topper’s name
s = {}

for i in range(5):
    k = input("\nEnter the Name :-")
    m = int(input("Enter the mark :-"))
    s[k] = m

high = max(s.values())

print("\nName of topper :-")
for i in s:
    if s[i] == high:
        print(i)
```

```python
#6 Bus Seat AllocationCreate a list of 10 seats. Let user input a seat number to book. Replace that index with “Booked”.
seats = ['Empty']*10

def seat_allo(seat_no):
    if seats[seat_no-1]=="Empty":
        seats[seat_no-1]="Booked"
        print(f"Your Seat is Booked of no {seat_no}")
    else:
        print(f"Seat no {seat_no} is already booked")
        

for i in range(int(input("How many seats you have to book"))):
    seat_allo(int(input("Enter seat no.")))
    
print(seats)
```

```python
#7 Vowel Counter Write a function that counts vowels in a given string.


def vowel():
    v="aeiou"
    string=input("Enter Sentance")
    c=0
    for i in string:
       if i in v:
           c+=1

    return c
       
print(vowel())
        
```

```python
#9 Password Strength Checker Input a password and check if it contains at least one uppercase, one lowercase, one digit, and one special character.
password = input("Enter your password")

length = len(password)
has_digit = 0
has_upper = 0
has_special = 0

for ch in password:
    if ch >= '0' and ch <= '9':
        has_digit = 1
    elif ch >= 'A' and ch <= 'Z':
        has_upper = 1
    elif not (ch >= 'a' and ch <= 'z'):
        has_special = 1

if length < 6:
    print("Weak Password")
elif has_digit and has_upper and has_special:
    print("Strong Password")
else:
    print("Medium Password")
```

```python
#10 Student Subject SelectionUse sets to show intersection and difference between students choosing Math and Science
def top_student(students):
    if not students:
        return None
        
    return max(students, key=lambda s: sum(s['marks']))


students_list = [
    {"name": "A", "marks": [90, 80, 75]},
    {"name": "B", "marks": [92, 88, 94]},
    {"name": "C", "marks": [70, 80, 75]},
    {"name": "D", "marks": [80, 60, 80]},
    {"name": "E", "marks": [75, 80, 80]}
]

topper = top_student(students_list)
print(f"Top Student: {topper['name']} with {sum(topper['marks'])} marks")
```
