```python
#1 Online Store Inventory Maintain a dictionary of items with quantity and price. 
#Allow user to buy items, update stock, and print final bill.

items = {
    "chocolate" : {"quantity" : 10, "price" : 20},
    "chips" : {"quantity" : 20, "price" : 10},
    "cookie" : {"quantity" : 14, "price" : 30},
    "wafer" : {"quantity" : 30, "price" : 10},
    "gum" : {"quantity" : 13, "price" : 10},
}

bought = {}

total = 0

while True:
    print("\n1. Buy Items")
    print("2. Final Bill")
    c = int(input("Enter your choice 1 or 2:-"))

    if c == 1:
        item = input("Enter the name of item :-")
        q = int(input("Enter the quantity :-"))

        if q > items[item]["quantity"]:
            print("\nStock is not available.")
        else:
            items[item]["quantity"] -= q
            total += q * items[item]["price"]
            if item not in bought:
                bought[item] = {"quantity" : q, "price" : items[item]["price"]}
            else:
                bought[item] = {"quantity" : bought[item]["quantity"] + q, "price" : items[item]["price"]}
            
            print("Purchase Done!!")

    elif c == 2:

        print("\nOrder Details :-")

        print(f" | {"Name" :<10} | {"Quantity":<10} | {"Price" :<10} | {"Total" :<10}")
        for i in bought:
            print(f" | {i :<10} | {bought[i]["quantity"] :<10} | {bought[i]["price"] :<10} | {bought[i]["quantity"] * bought[i]["price"] :<10}")

        print("\nFinal Bill :-",total)

        print("\nInventory Details :-")
        print(f" | {"Name" :<10} | {"Quantity":<10} | {"Price" :<10} |")
        for i in items:
            print(f" | {i :<10} | {items[i]["quantity"] :<10} | {items[i]["price"] :<10} | ")

        break

    else:
        print("\nError :- Enter valid number.")
```

```python
#2 # Movie Ratings Aggregator  Use a dictionary of {movie: [ratings]}. Write a function to calculate the average rating for each movie.

rating = {
    "Lion king" : [5.0, 4.0, 3.7, 4.5, 2.3],
    "zootopia" : [4.5, 3.5, 4.0, 3.4, 2.8],
    "coco" : [4.6, 3.5, 3.5, 4.5, 4.9],
    "toy story" : [4.8, 3.0, 4.4, 3.5, 4.8],
    "finding nemo" : [3.6, 4.2, 3.7, 4.6, 3.5]
}

ans = {k : round(sum(v)/len(v),1) for k,v in rating.items()}


print(f"{"Movie Name":<15}{"Average Rating"}")
for k,v in ans.items():
    print(f"{k : <15}{v}")
```

```python
# 3 Electricity Bill Generator Define a function that calculates bill amount using slabs (e.g., first 100 units ₹5/unit, next 100 ₹8/unit, etc.).

unit = int(input("Enter the units :-"))

if unit <= 100:
    total = unit * 5
else:
    total = (100 * 5) + (unit - 100) * 7


print("Unite :-",unit)
print("Total Bill :-",total)
```

```python
# 4 Word Frequency Counter  Use dictionary comprehension to count how many times each word occurs in a given paragraph.

text = input("Enter the text :-")

words = text.split()

c = {i : words.count(i) for i in words}

print(f"{"\nWords":<15}{"Frequency":<5}")

for i in c:
    print(f"{i:<15}{c[i]:<5}")
```

```python
#5 Email ID Formatter  Given a list of names, use map() and lambda to create email IDs in the format firstname.lastname@college.com.

names = ["Tom Doe", "jhon Smith", "Alice Johnson", "Bob Miller"]

email_ids = list(map(lambda name: f"{name.lower().split()[0]}.{name.lower().split()[1]}@college.com", names))

print("Generated Email IDs:")
for email in email_ids:
    print(f"- {email}")
```

```python
# 6 Duplicate Remover using Sets  Input a long list of items (with duplicates), use a set to remove duplicates and sort the result.

items = []
 
print("\nInsert 5 items :-")

for i in range(5):
    item = input(f"Enter the name of item - {i+1}. :-")
    items.append(item)

items = list(set(items))
items.sort()

print("\nSorted Items :- ",items) 
```

```python
#8 Student Report Card System Use nested dictionaries to store student → subjects → marks.
#Write a function to print each student’s average.

student = {
    "Samiya Mansur": {"Maths": 85, "Science": 90, "English": 78},
    "Jiya Patel": {"Maths": 92, "Science": 88, "English": 95}
}

def student_averages(records):
    """Calculates and prints the average marks for each student."""
    print("--- Student Average Marks ---")
    
    for student, subjects in records.items():
        marks = subjects.values()
        
        average = sum(marks) / len(marks)
        
        print(f"{student}: {average:.2f}")
    
    print("-----------------------------")

student_averages(student)
```

```python
#9Employee Salary Processor Use filter() to select employees with salary above a threshold,
#and reduce() to find total company salary expense.

from functools import reduce

e = {
    "Samiya" : 15000,
    "Jiya" : 12000,
    "Jay" : 40000,
    "Yash" : 80000
}

threshold = 20000
ans = list([i for i in e if e[i] in filter(lambda x : x > threshold, e.values())])
print("Employee's whose salary is above threshold :- ",ans)

expense = reduce(lambda x, y : x + y, e.values())
print("\nTotal company salary expense :-",expense)
```

```python
#10 Crypto Price Tracker Store crypto names and prices in a dictionary. 
#Write a function that returns the name of the cryptocurrency with the highest price.
crypto = {
    "BitCoin" : 120,
    "Ethereum" : 89,
    "Tether" : 78,
    "Solana" : 20,
    "Cardano" : 120,
}
def highest(a):
    high = max(a.values())
    return [i for i in a if a[i] == high]

for i in crypto:
    print(i,crypto[i])

print("\nHighest Price :-",highest(crypto))
```
