```python
#1 Café Bill Calculator Input the number of coffees, sandwiches, and pastries bought. Compute the total bill using fixed prices
c = int(input("Enter the number of coffees :-"))
s = int(input("Enter the number of sandwiches :-"))
p = int(input("Enter the number of pastries :-"))

total = c*100 + s*70 + p*150

print("\nOrder Details :-")
print("Coffees :-",c)
print("Sandwiches :-",s)
print("Pastries :-",p)
print("Total Bill :-",total)
```

```python
#2Temperature ConverterTake Celsius input and convert to Fahrenheit and Kelvin.

cel=eval(input("enter celcius to convert in ferenhit & Kelvin"))
def tempFern(cel):
    return cel*(9/5)+32

def tempKelwin(cel):
    return cel+273.15

print(cel,"celcius = ",tempFern(cel),"Ferenhit")
print(cel,"celcius = ",tempKelwin(cel),"Kelvin")
```

```python
#4 Attendance Warning System Input total classes and attended classes. 
#Print attendance percentage and whether the student is eligible for exams.

t = int(input("Enter total classes :-"))
a = int(input("Enter attended classes :-"))

p = round(a * 100 / t,2)

print("\nAttendance Percentage :-",p)
if p > 80:
    print("You are eligible for exam.")
else:
    print("You are not eligible for exam.")
```

```python
#5 Word Reverser Take a sentence and print the words in reverse order using string splitting and joining.

def word_reverser(sentance):
    mylist =sentance.split()
    mylist.reverse()
    return " ".join(mylist)
    

print(word_reverser("Python programming is fun!"))
```

```python
#7 Find the Largest of Three Numbers Using if-elif-else, print the largest among three numbers.

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
num3 = float(input("Enter third number: "))

if (num1 >= num2) and (num1 >= num3):
    largest = num1
elif (num2 >= num1) and (num2 >= num3):
    largest = num2
else:
    largest = num3

print(f"The largest number is: {largest}")

```

```python
#8Leap Year Checker Input a year and check whether it’s a leap year
import calendar as c

y = int(input("Enter the year :-"))

if c.isleap(y):
    print(y,"is leap year.")
else:
    print(y,"is not leap year.")
```

```python
#9 Shopping Discount System If the purchase amount > 5000, give 10% discount; else, print “No Discount”.
purchase_amount = float(input("Enter the total purchase amount: "))

if purchase_amount > 5000:
    discount = purchase_amount * 0.10
    final_amount = purchase_amount - discount
    print(f"10% Discount Applied! Saved: {discount:.2f}")
    print(f"Final Amount to Pay: {final_amount:.2f}")
else:
    print("No Discount")
    print(f"Final Amount to Pay: {purchase_amount:.2f}")
```
