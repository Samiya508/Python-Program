```python
drop_index=df.loc[(df['x']==0) | (df['y']==0) | (df['z']==0) ].index
df=df.drop(drop_index)
df.describe
```

```python
print(df['cut'].unique())
```

```python
df['cut']=df['cut'].factorize()[0]
df['color']=df['color'].factorize()[0]
df['clarity']=df['clarity'].factorize()[0]
df.head()
```

```python
#which diamond is best and which to buy?

import seaborn as sns

buy = df.copy()
buy["price_per_carat"] = buy["price"] / buy["carat"]
buy = buy.sort_values(by="price_per_carat", ascending=True)

print(f"Best value for money: {round(min(buy['price_per_carat']), 2)}$")
print("The first 3:\n")


plt.figure(figsize=(6, 4))
sns.countplot(x="cut", data=df, color="green")
plt.title("Distribution of Cut")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x="color", data=df, color="green")
plt.title("Distribution of Color")
plt.show()

plt.figure(figsize=(6, 4))
sns.countplot(x="clarity", data=df, color="green")
plt.title("Distribution of Clarity")
plt.show()

plt.figure(figsize=(6, 6))
cut_counts = df["cut"].value_counts()

plt.pie(cut_counts, labels=cut_counts.index, autopct="%1.1f%%")
plt.title("Pie Chart of Cut")
plt.show()


plt.figure(figsize=(6, 6))
color_counts = df["color"].value_counts()
plt.pie(color_counts, labels=color_counts.index, autopct="%1.1f%%")
plt.title("Pie Chart of Color")
plt.show()
```
