```python
import random
import time

def math_magic_game():
   
    random_num = random.randint(1,50)
    result = random_num / 2
    
    print("--- Welcome to the Math Mind Reader ---")
    print("Follow the instructions carefully in your head.\n")

    
    input("1. Think of a secret number in your mind. (Press Enter when ready)")

    
    input("2. Multiply your secret number by 2. (Press Enter when ready)")

    
    print("3. Now, add", random_num, "to your total.")
    input("   (Press Enter when ready)")

   
    input("4. Divide the new total by 2. (Press Enter when ready)")

    
    input("5. Finally, subtract your original secret number from the total. (Press Enter when ready)")

    
    print("\nCalculating your final result...")
    time.sleep(2)
    
    print("The number you are thinking of right now is:",result,"!")
    print("---------------------------------------")

math_magic_game()
```
