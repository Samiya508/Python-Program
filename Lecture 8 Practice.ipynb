```python
quality_counts=wine_df['quality'].value_counts()

plt.figure(figsize=[8,6])
plt.bar(quality_counts.index,quality_counts,color='pink')
plt.title("Count plot of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
quality_counts=wine_df['quality'].value_counts()

plt.figure(figsize=(10, 8))
sns.swarmplot(x="quality", y="alcohol", data=wine_df, palette='viridis')
plt.title('Swarm Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
quality_counts=wine_df['quality'].astype(str)

wine_df['quality'] = wine_df['quality'].astype(str)  # Convert 'quality' to categorical
plt.figure(figsize=(10, 8))
sns.violinplot(x="quality", y="alcohol", data=wine_df, palette={
               '3': 'lightcoral', '4': 'lightblue', '5': 'lightgreen', '6': 'gold', '7': 'lightskyblue', '8': 'lightpink'}, alpha=0.7)
plt.title('Violin Plot for Quality and Alcohol')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
sns.boxplot(x='quality',y='alcohol',data=wine_df)
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(wine_df.corr(),annot=True,fmt='.2f',cmap='Pastel2',linewidths=2)
plt.title('Correlarion Heatmap')
plt.show()
```

```python
duplicated=wine_df.duplicated()
print("number of duplicate:",duplicated.sum())
wine_df[duplicated]
```

```python
wine_df=wine_df.drop_duplicates()
duplicated=wine_df.duplicated()
print("number of duplicated instances:",duplicated.sum())
wine_df[duplicated]
```

```python
wine_df["quality"]=np.where(wine_df["quality"]>=7,1,0)
wine_df["quality"].astype(int)
wine_df["quality"].value_counts()
```

```python
features = wine_df.columns[:-1].values
label = [wine_df.columns[-1]]

print ("The Features are:", features)
print ("The Label is:", label)
```

```python
# Set custom labels
labels = ["3", "4", "5", "6", "7", "8"]
ticks = range(len(labels))
# Create a figure
fig, ax = plt.subplots(figsize=(8, 5))
# Plot the count plot
sns.countplot(data=wine_df, x='quality', ax=ax, palette=sns.color_palette("BuPu"))
ax.set_title('Distribution of Red Wine Quality')
ax.set_xticks(ticks)
ax.set_xticklabels(labels)

# Adjust layout to prevent overlapping
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px

# Quality Distribution
content_type = wine_df.groupby(['quality']).size().reset_index(name='counts')
# Map the quality of red wine to specific labels
content_type['quality_label'] = content_type['quality'].map(dict(zip(range(3, 9), labels)))
# Plot the red wine quality distribution
fig = px.pie(content_type, 
             names='quality_label', 
             values='counts', 
             color_discrete_sequence=px.colors.sequential.Agsunset,
             title='Distribution of Red Wine Quality', 
             height=500, 
             width=700)
# Include legend and labels
fig.update_layout(
    legend_title="Quality",
    legend=dict(
        x=0.8,
        y=1.0
    )
)
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.set(style="whitegrid")
ax = sns.barplot(x="quality", y="alcohol", data=wine_df)
plt.title("Most Positive Relationship", fontsize=16)
ax.set_xlabel("quality", fontsize=12)
ax.set_ylabel("alcohol", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Create line plots for 'alcohol' and 'residual sugar' against 'density'
sns.lineplot(x="alcohol", y="density", data=wine_df, label='alcohol', marker='o')
sns.lineplot(x="residual sugar", y="density", data=wine_df, label='residual Sugar', marker='s')

plt.title("Density vs. Alcohol and Residual Sugar", fontsize=16)
plt.xlabel("Density", fontsize=12)
plt.ylabel("Value", fontsize=12)
plt.legend()
plt.tight_layout()
plt.show()
 
```
