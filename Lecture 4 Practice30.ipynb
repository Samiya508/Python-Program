```python
#3 Create a generator function that yields the Fibonacci sequence up to n terms.
def fibonacci_generator(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b

fib_sequence = list(fibonacci_generator(10))
print(fib_sequence)
```

```python
                #HARD Program
# 1 Write a generator that produces an infinite sequence of prime numbers; use next() to get the first 10 primes.

def prime_generator():
    yield 2
    primes = [2]
    n = 3
    while True:
        is_prime = True
        for p in primes:
            if p * p > n:
                break
            if n % p == 0:
                is_prime = False
                break
        
        if is_prime:
            primes.append(n)
            yield n
        n += 2

gen = prime_generator()

first_10 = [next(gen) for _ in range(10)]

print(first_10)
```

```python
# 3Write a function that accepts a list of dictionaries (each representing a student with marks) and returns the student with the highest total marks.
def get_top_student(students):
    if not students:
        return None
        
    return max(students, key=lambda s: sum(s['marks']))


students_list = [
    {"name": "A", "marks": [85, 90, 88]},
    {"name": "B", "marks": [92, 88, 94]},
    {"name": "C", "marks": [70, 80, 75]}
]

top_student = get_top_student(students_list)
print(f"Top Student: {top_student['name']} with {sum(top_student['marks'])} marks")
```

```python
# 4 Use map() and a lambda function to normalize a list of numbers between 0 and 1.
numbers = [10, 20, 30, 40, 50]

min_val = min(numbers)
max_val = max(numbers)
diff = max_val - min_val

normalized = list(map(lambda x: (x - min_val) / diff, numbers))

print(normalized)
```

```python
# 6 Write a generator to yield all palindromic numbers between 1 and 1000.
def palindrome_generator(limit):
    for n in range(1, limit + 1):
        s = str(n)
        if s == s[::-1]:
            yield n
palindromes = list(palindrome_generator(1000))
print(palindromes)
```

```python
# 7 Write a function that accepts another function as a parameter and applies it to a list of numbers (demonstrating higher-order functions).

def apply_operation(func, numbers):
    return [func(n) for n in numbers]

# Example functions to pass in
def square(x):
    return x ** 2

def cube(x):
    return x ** 3

nums = [1, 2, 3, 4, 5]

squared_nums = apply_operation(square, nums)
cubed_nums = apply_operation(cube, nums)

print(f"Original: {nums}")
print(f"Squared:  {squared_nums}")
print(f"Cubed:    {cubed_nums}")
```

```python
# 9 Write a recursive function to flatten a nested dictionary (keys can be nested dictionaries) into a single-level dictionary with tuple keys representing the path.
def flatten_dict(nested_dict, path=()):
    items = {}
    for key, value in nested_dict.items():
        new_path = path + (key,)
        if isinstance(value, dict):
            items.update(flatten_dict(value, new_path))
        else:
            items[new_path] = value
    return items

nested = {
    "user": {
        "id": 1,
        "profile": {
            "name": "Alice",
            "city": "NY"
        }
    },
    "status": "active"
}

flat = flatten_dict(nested)
print(flat)
```

```python
# 10 Write a function to calculate the factorial of a number using a generator to yield intermediate factorial values, and use reduce() to compute the final result.
from functools import reduce

def factorial_steps(n):
    for i in range(1, n + 1):
        yield i

def calculate_factorial(n):
    if n == 0:
        return 1
    return reduce(lambda x, y: x * y, factorial_steps(n))

num = 5
result = calculate_factorial(num)
print(f"The factorial of {num} is {result}")
```
